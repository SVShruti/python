```python
import pandas as pd
import matplotlib.pyplot as plt
x= pd.read_csv("/content/diabetes.tab.txt", sep='\t')
x
```

```python
#summary of data
x.info()
```

```python
#statistical value

x.describe()
```

```python
#calculating mean of BMI
mean_bmi= x['BMI'].mean()
min_bmi= x['BMI'].min()
max_bmi= x['BMI'].max()

print(mean_bmi)
print(min_bmi)
print(max_bmi)
```

```python
#calculating mean, std, min and max of BP

mean_bp=x['BP'].mean()
std_bp=x['BP'].std()
min_bp=x['BP'].min()
max_bp=x['BP'].max()

print("Mean BP is: " +str(mean_bp))
print("Standard Deviation BP is: " +str(std_bp))
print("Minimum BP is: " +str(min_bp))
print("Maximum BP is: " +str(max_bp))

```

```python
#Visualizations
plt.figure(figsize=(8,2))
plt.boxplot(x["BMI"],vert=False, showmeans= True)
plt.title("This is a Boxplot for BMI of patients ")
plt.xlabel("BMI of patients ")
plt.grid()
plt.show()
```

```python
#Age

plt.figure(figsize=(8,2))
plt.boxplot(x['AGE'],vert=False, showmeans= True)
plt.title("This is a Boxplot for Age of patients ")
plt.xlabel("Age of patients ")
plt.grid()
plt.show()
```

```python
plt.hist(x['AGE'], bins=10)
plt.title("This is a Histogram for Age of patients ")
plt.xlabel("Age of patients ")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
plt.show()
```

```python
#making a pie chart for age and sex

plt.pie(x["SEX"])
plt.show()
```
